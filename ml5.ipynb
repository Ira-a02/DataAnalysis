```python
import pandas as pd
from sklearn import linear_model
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import Perceptron
import numpy as np
```

 1. Загрузите обучающую и тестовую выборки из файлов perceptron
train.csv и perceptron-test.csv. Целевая переменная записана в первом столбце, признаки во втором и третьем

```python
df_test = pd.read_csv('perceptron-test.csv')
df_train = pd.read_csv('perceptron-train.csv')

X_test=np.array(df_test.iloc[:,1:3].values)
y_test=np.array(df_test.iloc[:, 0].values)
X_train=np.array(df_train.iloc[:,1:3].values)
y_train=np.array(df_train.iloc[:, 0].values)
```

 2. Обучите персептрон со стандартными параметрами и random_state=241.

```python
clf=Perceptron(random_state=241)
clf.fit(X_train,y_train)
predictions= clf.predict(X_test)
```

 3. Подсчитайте качество (долю правильно классифицированных объ
ектов, accuracy) полученного классификатора на тестовой выборке

```python
accuracy = accuracy_score(y_test, predictions)
```

 4. Нормализуйте обучающую и тестовую выборку с помощью класса
 StandardScaler

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

 5. Обучите персептрон на новых выборках. Найдите долю правиль
ных ответов на тестовой выборке

```python
clf_st=Perceptron(random_state=241)
clf_st.fit(X_train_scaled,y_train)
predictions_st= clf_st.predict(X_test_scaled)

accuracy_st = accuracy_score(y_test, predictions_st)
```

 6. Найдите разность между качеством на тестовой выборке после нормализации и качеством до нее. Это число и будет ответом на задание

```python
accuracy_difference = accuracy_st - accuracy

with open('accuracy_difference.txt', 'w') as f:
    f.write(f"{round(accuracy_difference,3)}")

print("Разность между качеством на тестовой выборке:", accuracy_difference)
```
